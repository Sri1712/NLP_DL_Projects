### Prediction of the ANN trained model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import load_model
import pickle
```

```python
# Load the trained model, scaler, and one-hot label encoder
model = load_model('model.h5')

with open('gender_encoder.pkl', 'rb') as file:
    gender_encoder = pickle.load(file)

with open('geo_onehot_encoder.pkl', 'rb') as file:
    geo_onehot_encoder = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'Germany',
    'Gender': 'Female',
    'Age': 46,
    'Tenure': 3,
    'Balance': 30000,
    'NumOfProducts': 3,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 40000
}
```

```python
# One-hot encode 'Geography'
geo_encoded = geo_onehot_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=geo_onehot_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
## Encode categorical variables
input_df['Gender']=gender_encoder.transform(input_df['Gender'])
input_df
```

```python
## concatination one hot encoded 
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_df
```

```python
## Scaling the input data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
# Make prediction
prediction=model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
print(prediction_proba)
```

```python
if prediction_proba > 0.5:
    print('The customer is likely to churn.')
else:
    print('The customer is not likely to churn.')
```
