##### **Simple RNN**

```python
from tensorflow.keras.preprocessing.text import one_hot, Tokenizer
```

```python
sentence = ["The cat is on the table",
            "The dog is under the table",
            "The cat is under the table",
            "The rainbow is in the sky",
            "The umbrella is important on a rainy day",
            "I love to eat pizza on a rainy day",
            "The cat is on the roof",
            "The wonderful world is full of surprises"]

vocab_size = 10000
```

```python
sentence
```

```python
## One hot representation
onehot_repr = [one_hot(words, vocab_size) for words in sentence]
onehot_repr
```

```python
# Tokenizer
vocab_size = 10000
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(sentence)
sequences = tokenizer.texts_to_sequences(sentence)
```

```python
# Word embedding representation
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=10
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)

# Feature representation
embedding_vector_features=40
```

```python
# Create the model
model=Sequential()
model.add(Embedding(input_dim=vocab_size,output_dim=embedding_vector_features,input_length=sent_length))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
